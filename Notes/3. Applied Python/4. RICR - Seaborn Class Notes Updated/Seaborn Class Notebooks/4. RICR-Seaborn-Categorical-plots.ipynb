# Categorical Plots

#### Figure level function: catplot()
#### Axis level functions: 
1. stripplot 
2. swarmplot
3. boxplot
4. violinplot
5. barplot
6. countplot
7. pointplot

```python
import seaborn as sns
```

```python
df= sns.load_dataset("penguins")
```

```python
df.head()
```

## Categorical scatter plot
1. stripplot
2. swarmplot

```python
# 1. 
sns.stripplot(data=df, x='species', y='body_mass_g')
```

```python
# jitter
sns.stripplot(data=df, x='species', y='body_mass_g',jitter=False)
```

```python
# hue
sns.stripplot(data=df, x='species', y='body_mass_g',hue='sex')
```

```python
# figure level function -> catplot
sns.catplot(data=df, x='species', y='flipper_length_mm',kind="strip")
```

```python
sns.catplot(data=df, x='species', y='flipper_length_mm',kind="strip", jitter=0.3)
```

```python
# hue
sns.catplot(data=df, x='species', y='flipper_length_mm',kind="strip",hue='sex')
```

2. swarmplot -> shows the distribution of data more clear 

```python
sns.swarmplot(data=df,x='species',y='body_mass_g')
```

```python
sns.catplot(data=df,x='species',y='body_mass_g',kind='swarm')
```

```python
sns.catplot(data=df,x='species',y='body_mass_g',kind='swarm',hue='sex')
```

## Categorical Distribution plot
1. Boxplot
2. Violinplot


```python
sns.boxplot(data=df,x='sex',y='body_mass_g')
```

```python
sns.boxplot(data=df,x='species',y='body_mass_g')
```

```python
#hue
sns.boxplot(data=df,x='species',y='body_mass_g',hue='sex')
```

```python
# univariate analysis using boxplot
sns.boxplot(data=df,y='flipper_length_mm')
```

```python
# using figure level function
sns.catplot(data=df, x='species', y='body_mass_g',hue='sex',kind='box')
```

2. Violinplot -> (boxplot+KDEplot)

```python
sns.violinplot(data=df,x='species',y='flipper_length_mm')
```

```python
# hue
sns.violinplot(data=df,x='species',y='body_mass_g',hue='sex')
```

```python
# split
sns.violinplot(data=df,x='species',y='body_mass_g',hue='sex',split=True)
```

```python
# violinplot using figure level catplot
sns.catplot(data=df,x='species',y='body_mass_g',hue='sex',split=True,kind='violin')
```

## Categorical estimate plots
1. barplot
2. pointplot
2. countplot

1. barplot()

-> The upper black lines are known as errorbar

```python
sns.barplot(data=df,x='species',y='body_mass_g')
```

```python
# hue
sns.barplot(data=df,x='species',y='body_mass_g',hue='sex')
```

```python
# errorbar
sns.barplot(data=df,x='species',y='body_mass_g',hue='sex',errorbar=None)
```

```python
# figure level of barplot
sns.catplot(data=df,x='species',y='body_mass_g',hue='sex',kind='bar')
```

2. pointplot()    
-> It just remove the bars from bar plot and connect the top most points of the bars.

```python
sns.pointplot(data=df,x='species',y='flipper_length_mm')
```

errorbar = None
-> It remove the errorbars

```python
sns.pointplot(data=df,x='species',y='flipper_length_mm',errorbar=None)
```

```python
# figure level of pointplot
sns.catplot(data=df,x='species',y='body_mass_g',hue='sex',errorbar=None,kind='point')
```

### 3. countplot()

```python
sns.countplot(data=df,x='island')
```

```python
# hue
sns.countplot(data=df,x='island',hue='species')
```

```python
# figure level of countplot()
sns.catplot(data=df,x='island',hue='species',kind='count')
```

## faceting using catplot()   
-> more than one graph in a single figure.

```python
sns.catplot(data=df,x='species',y='body_mass_g',kind='box',col='sex',row='island')
```
