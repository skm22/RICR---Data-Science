## Seaborn: Matrix Plots

### What are Matrix Plots?

- **Matrix Plots**: Visualizations that show data in a grid format.
- **Each cell in the grid** represents a data point.

### Why Use Matrix Plots?

- To easily see **patterns, relationships, and anomalies** in data.
- Useful for **comparing multiple variables**.

### What Do They Show?

- **Heatmaps**: Color-coded representation of data.
- **Cluster Maps**: Heatmaps with hierarchical clustering.

### When to Use Matrix Plots?

- When you need to **visualize relationships** between multiple variables.
- For **correlation matrices**, **feature comparisons**, and **pattern recognition**.


```python
import seaborn as sns
```

## Type of data :
1. Numerical data
2. Categorical data.

```python
df = sns.load_dataset('penguins')
```

```python
df.sample(6)
```

## Heatmap()

```python
# Calculate correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Heatmap
sns.heatmap(correlation_matrix)
```

## Parameters ->
1. annot = True

```python
sns.heatmap(correlation_matrix , annot = True)
```

## Parameter ->
4. fmt=".1f",  Format values to two decimal places

```python
sns.heatmap(correlation_matrix , annot = True, fmt = '.1f')
```

## Parameter->
2. linewidth and linecolor.

```python
sns.heatmap(correlation_matrix , annot = True, linewidths=2, linecolor="m" )
```

## Parameter ->
3. cmap = 'viridis' or 'summer' or 'winter' or 'jet' or 'coolwarm'

```python
sns.heatmap(correlation_matrix,annot = True,linewidth = 2,cmap = 'coolwarm')
```

## clustermap()

```python
df
```

```python
# Prepare the data for clustering
# Selecting numerical columns only
numeric_data = df.select_dtypes(include=["float", "int"]).dropna()

# Clustermap
sns.clustermap(
    numeric_data.corr(),  # Correlation matrix as input
    method="complete",  # Hierarchical clustering method (e.g., single, complete, average)
    cmap="coolwarm",  # Color map
    annot=True,  # Show values in cells
    figsize=(10, 8)  # Adjust figure size
)
```
