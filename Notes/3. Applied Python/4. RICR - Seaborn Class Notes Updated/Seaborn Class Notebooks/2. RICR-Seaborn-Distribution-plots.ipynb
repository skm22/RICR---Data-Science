## Properties.

1. Used for univariate anaylsis.
2. Used to find out the distribution
3. Range of the observation
4. Central tendency

### What is shows?
1. Is the data bimodel?
2. Are their outliers.

```python
import seaborn as sns
```

## Two functions
## Figure level functions -> displot
## Axes level functions -> histplot , rugplot , kdeplot.

```python
df = sns.load_dataset('penguins')
```

```python
df.head()
```

## 1. histplot

a. Axes level function

```python
sns.histplot(data=df , x = 'body_mass_g')
```

figure level function -> displot for (histplot)

```python
sns.displot(kind = 'hist' , data = df , x = 'body_mass_g' , bins = 20)
```

### Histplot on categorical data.
 - Act as a countplot

```python
sns.displot(kind = 'hist' , data = df , x = 'species')
```

### hue parameter -> in histplot

```python
sns.displot(kind = 'hist' , data = df , x = 'flipper_length_mm' , hue = 'species')
```

### Element parameter -> in histplot

element can be : bars, step, poly

```python
sns.displot(kind = 'hist' , data = df , x = 'flipper_length_mm' , hue = 'species' , element = 'poly')
```

### Bivariate analysis using histplot.

```python
sns.histplot(data = df , x = 'flipper_length_mm' , y = 'bill_length_mm')
```

## Facet plots of histplot -->

```python
sns.displot(kind = 'hist' , data = df , x = 'flipper_length_mm', hue='species' , col = 'sex')
```

# kdeplot
#### Rather than using discrete bins , a KDE plot smooths the observations with a gaussian kernal, producing a continous density estimate.

```python
df.head()
```

a. Axes level function.

```python
sns.kdeplot(data = df , x = 'flipper_length_mm')
```

b. figure level function -> displot for (kde)

```python
sns.displot(kind = 'kde' , data = df , x = 'flipper_length_mm')
```

### Fill parameter.

```python
sns.displot(kind = 'kde' , data = df , x = 'flipper_length_mm' , hue = 'sex' , fill = True)
```

### height parameter.

```python
sns.displot(kind = 'kde' , data = df , x = 'flipper_length_mm' , hue = 'sex' , fill = True , height = 7)
```

## Bivariate using kdeplot.

```python
sns.kdeplot(data = df , x = 'flipper_length_mm' , y = 'bill_length_mm')
```

## rugplot -> supporting plot..

```python
sns.kdeplot(data = df , x = 'flipper_length_mm')
sns.rugplot(data = df , x = 'flipper_length_mm')
```
