```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(4)
```

```python
titanic.info()
```

```python
# Scatterplot: Relationship between age and fare
sns.relplot(data=titanic, x='age', y='fare', hue='class', style='sex')
plt.title('Scatterplot of Age vs Fare')
plt.show()
```

```python
# Lineplot: Fare changes with age, separated by class
sns.lineplot(data=titanic, x='age', y='fare', hue='class', style='sex', markers=True)
plt.title('Lineplot of Age vs Fare')
plt.show()
```

```python
# Histplot: Distribution of passenger ages
sns.histplot(data=titanic, x='age', kde=True)
plt.title('Histogram of Passenger Ages')
plt.show()
```

```python
# KDEplot: Distribution of fares by class
sns.kdeplot(data=titanic, x='fare', hue='class', fill=True)
plt.title('KDE Plot of Fares by Class')
plt.show()
```

```python
# Barplot: Average fare by class and sex
sns.barplot(data=titanic, x='class', y='fare', hue='sex')
plt.title('Barplot of Average Fare by Class')
plt.show()
```

```python
# Countplot: Count of passengers by class
sns.countplot(data=titanic, x='class', hue='sex')
plt.title('Countplot of Passengers by Class')
plt.show()
```

```python
# Boxplot: Distribution of ages by class
sns.boxplot(data=titanic, x='class', y='age', hue='sex')
plt.title('Boxplot of Ages by Class')
plt.show()
```

```python
# Violinplot: Distribution of ages by class
sns.violinplot(data=titanic, x='class', y='age', hue='sex', split=True)
plt.title('Violinplot of Ages by Class')
plt.show()
```

```python
# Stripplot: Fare by class
sns.stripplot(data=titanic, x='class', y='fare', hue='sex', jitter=True)
plt.title('Stripplot of Fare by Class')
plt.show()
```

```python
numeric_titanic = titanic.select_dtypes(include=['number'])
```

```python
# Heatmap: Correlation between numerical variables
sns.heatmap(numeric_titanic.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlations - Titanic Dataset')
plt.show()
```

```python
# Clustermap: Clustered heatmap of correlations
sns.clustermap(numeric_titanic.corr(), annot=True, cmap='coolwarm')
plt.title('Clustermap of Correlations - Titanic Dataset')
plt.show()
```

```python
# Pairplot: Pairwise relationships in the dataset
sns.pairplot(titanic.select_dtypes(include=['number']), hue='survived')
plt.title('Pairplot of Titanic Dataset')
plt.show()
```

```python
# Jointplot: Scatter plot with marginal distributions
sns.jointplot(data=titanic, x='age', y='fare', kind='scatter')
plt.title('Jointplot of Age vs Fare')
plt.show()
```

```python
# Jointplot: KDE with marginal distributions
sns.jointplot(data=titanic, x='age', y='fare', kind='kde')
plt.title('Jointplot of Age vs Fare (KDE)')
plt.show()
```

```python
# FacetGrid: Age vs fare, faceted by class and survived
g = sns.FacetGrid(titanic, col="class", row="survived")
g.map(sns.scatterplot, "age", "fare")
plt.show()
```
