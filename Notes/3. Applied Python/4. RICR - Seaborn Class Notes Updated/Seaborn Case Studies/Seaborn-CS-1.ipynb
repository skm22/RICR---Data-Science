```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# Scatterplot: Relationship between total bill and tip
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
plt.title('Scatterplot of Total Bill vs Tip')
plt.show()
```

```python
# Scatterplot: Relationship between total bill and tip
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
plt.title('Scatterplot of Total Bill vs Tip')
plt.show()
```

```python
tips.columns
```

```python
Sun_tips = tips[tips['day'] == "Sun"]
```

```python
# Lineplot: Tip amount changes with total bill for each day
sns.lineplot(data=Sun_tips, x='total_bill', y='tip', hue='day', style='time', markers=True)
plt.title('Lineplot of Total Bill vs Tip')
plt.show()
```

```python
# Histplot: Distribution of total bills
sns.histplot(data=tips, x='total_bill', kde=True)
plt.title('Histogram of Total Bill')
plt.show()
```

```python
# KDEplot: Distribution of tips by gender
sns.kdeplot(data=tips, x='tip', hue='sex', fill=True)
plt.title('KDE Plot of Tips by Gender')
plt.show()
```

```python
# Barplot: Average tip by day
sns.barplot(data=tips, x='day', y='tip', hue='sex')
plt.title('Barplot of Average Tip by Day')
plt.show()
```

```python
# Countplot: Count of meals by day
sns.countplot(data=tips, x='day', hue='sex')
plt.title('Countplot of Meals by Day')
plt.show()
```

```python
# Boxplot: Distribution of total bills by day
sns.boxplot(data=tips, x='day', y='total_bill', hue='time')
plt.title('Boxplot of Total Bills by Day')
plt.show()
```

```python
# Violinplot: Distribution of total bills by day
sns.violinplot(data=tips, x='day', y='total_bill', hue='time', split=True)
plt.title('Violinplot of Total Bills by Day')
plt.show()
```

```python
# Stripplot: Tips by day
sns.stripplot(data=tips, x='day', y='tip', hue='sex', jitter=True)
plt.title('Stripplot of Tips by Day')
plt.show()
```

```python
# Swarmplot: Tips by day
sns.swarmplot(data=tips, x='day', y='tip', hue='sex')
plt.title('Swarmplot of Tips by Day')
plt.show()
```

```python
# Select only numeric columns for correlation
numeric_tips = tips.select_dtypes(include=['number'])
```

```python

# Now create the heatmap
sns.heatmap(numeric_tips.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlations')
plt.show()
```

```python
# Clustermap: Clustered heatmap of correlations
sns.clustermap(numeric_tips.corr(), annot=True, cmap='coolwarm')
plt.title('Clustermap of Correlations')
plt.show()
```

```python
# Pairplot: Pairwise relationships in the dataset
sns.pairplot(tips, hue='sex')
plt.title('Pairplot of Tips Dataset')
plt.show()
```

```python
# Jointplot: Scatter plot with marginal distributions
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter')
plt.title('Jointplot of Total Bill vs Tip')
plt.show()
```

```python
# Jointplot: KDE with marginal distributions
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
plt.title('Jointplot of Total Bill vs Tip (KDE)')
plt.show()
```
