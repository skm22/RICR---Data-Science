```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\hp\Desktop\Datasets\Dataset Samples\StudentsPerformance.csv")
```

```python
sns.relplot(data=data, x='writing score', y='reading score', hue='gender', style='lunch')
plt.title('Scatterplot of Writing Score vs Reading Score')
plt.show()
```

```python
sns.lineplot(data=data, x='writing score', y='reading score', hue='gender', style='lunch', markers=True)
plt.title('Lineplot of Writing Score vs Reading Score')
plt.show()
```

```python
sns.histplot(data=data, x='math score', kde=True, hue='gender')
plt.title('Histogram of Math Scores')
plt.show()
```

```python
sns.kdeplot(data=data, x='reading score', hue='gender', fill=True)
plt.title('KDE Plot of Reading Scores by Gender')
plt.show()
```

```python
sns.barplot(data=data, x='parental level of education', y='math score', hue='gender')
plt.title('Barplot of Math Score by Parental Education Level')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=data, x='test preparation course', hue='gender')
plt.title('Countplot of Test Preparation Course by Gender')
plt.show()
```

```python
sns.boxplot(data=data, x='lunch', y='reading score', hue='gender')
plt.title('Boxplot of Reading Scores by Lunch Type')
plt.show()
```

```python
sns.violinplot(data=data, x='lunch', y='writing score', hue='gender', split=True)
plt.title('Violinplot of Writing Scores by Lunch Type')
plt.show()
```

```python
# First, convert categorical data to numerical format if necessary
numeric_data = data.select_dtypes(include=['number'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlations')
plt.show()
```

```python
sns.clustermap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Clustermap of Correlations')
plt.show()
```

```python
sns.pairplot(data.select_dtypes(include=['number']))
plt.title('Pairplot of Student Performance')
plt.show()
```

```python
sns.jointplot(data=data, x='math score', y='writing score', kind='scatter')
plt.title('Jointplot of Math Score vs Writing Score')
plt.show()
```
