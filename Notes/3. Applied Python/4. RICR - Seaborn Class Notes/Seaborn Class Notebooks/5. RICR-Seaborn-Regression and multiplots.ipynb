# Regression plot
-> primarily intended to add a visual guide that helps to emphasize patterns in a dataset during exploratory data analyses   
1. regplot() -> axis level function
2. lmplot() -> figure level function

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

1. regplot()

```python
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
# hue parameter is not present in regplot.
sns.regplot(data=tips,x='total_bill',y='tip',hue='sex') #Error
```

2. lmplot()

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue="sex")
```

-> residual plot -> calculating error in regression model.

```python
#residplot()
sns.residplot(data=tips,x='total_bill',y='tip')
```

# Multigraph

1. pairplot / pairgrid
2. jointplot / jointgrid

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

1. pairplot()

```python
sns.pairplot(iris)
```

```python
# hue
sns.pairplot(iris,hue='species')
```

```python
# pairgrid()
a = sns.PairGrid(iris,hue='species')
a.map(sns.scatterplot)
```

-> In PairGrid you have the customization that you can change any plot type

```python
# map_offdiag ans map_diag

a = sns.PairGrid(iris,hue='species')
a.map_offdiag(sns.scatterplot)
a.map_diag(sns.histplot)
```

```python
a = sns.PairGrid(iris,hue='species')
a.map_offdiag(sns.histplot)
a.map_diag(sns.violinplot)
```

```python
# map_upper and map_lower
a = sns.PairGrid(iris,hue='species')
a.map_diag(sns.histplot)
a.map_upper(sns.scatterplot)
a.map_lower(sns.kdeplot)
```

```python
# vars
a = sns.PairGrid(iris,hue='species',vars=['sepal_width','petal_width'])
a.map_diag(sns.histplot)
a.map_upper(sns.scatterplot)
a.map_lower(sns.kdeplot)
```

2. jointplot

```python
# kind = scatter
sns.jointplot(data=tips,x='total_bill',y='tip',kind='scatter')
```

```python
# kind = kde
sns.jointplot(data=tips,x='total_bill',y='tip',kind='kde')
```

```python
# kind =hist
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist')
```

```python
# hue
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist',hue='sex')
```

JointGrid()

```python
a = sns.JointGrid(tips,x='total_bill',y='tip')
a.plot(sns.scatterplot,sns.boxplot)
```

```python
a = sns.JointGrid(tips,x='total_bill',y='tip')
a.plot(sns.kdeplot,sns.histplot)
```
