# Distribution Plots

Screenshot 2024-07-10 175312.png

## Properties.

1. Used for univariate anaylsis.
2. Used to find out the distribution
3. Range of the observation
4. Central tendency

### What is shows?
1. Is the data bimodel?
2. Are their outliers.

```python
import seaborn as sns
```

## Two functions
## Figure level functions -> displot
## Axes level functions -> histplot , rugplot , kdeplot.

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.histplot(data=df , x = 'total_bill')
```

```python
sns.displot(kind = 'hist' , data = df , x = 'total_bill' , bins = 20)
```

## Histplot on categorical data.
 - Act as a countplot

```python
sns.displot(kind = 'hist' , data = df , x = 'day')
```

## hue parameter -> in histplot

```python
sns.displot(kind = 'hist' , data = df , x = 'tip' , hue = 'sex')
```

## Element parameter -> in histplot

```python
sns.displot(kind = 'hist' , data = df , x = 'tip' , hue = 'sex' , element = 'step')
```

## Bivariate analysis using histplot.

```python
sns.histplot(data = df , x = 'total_bill' , y = 'tip')
```

## Facet plots of histplot -->

```python
sns.displot(kind = 'hist' , data = df , x = 'tip' , col = 'sex')
```

## kdeplot -> axes level function.
## Rather than using discrete bins , a KDE plot smooths the observations with a gaussian kernal, producing a continous density estimate.

```python
df.head()
```

```python
sns.kdeplot(data = df , x = 'total_bill')
```

```python
sns.displot(kind = 'kde' , data = df , x = 'total_bill')
```

```python
sns.displot(kind = 'kde' , data = df , x = 'total_bill' , hue = 'sex' , fill = True)
```

```python
sns.displot(kind = 'kde' , data = df , x = 'total_bill' , hue = 'sex' , fill = True , height = 10)
```

## Bivariate using kdeplot.

```python
sns.kdeplot(data = df , x = 'total_bill' , y = 'tip')
```

## rugplot -> supporting plot..

```python
sns.kdeplot(data = df , x = 'total_bill')
sns.rugplot(data = df , x = 'total_bill')
```
