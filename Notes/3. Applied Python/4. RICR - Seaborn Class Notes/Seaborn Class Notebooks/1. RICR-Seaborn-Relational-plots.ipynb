## What is Seaborn?

- **Seaborn**: A Python library for data visualization, built on top of Matplotlib.
- **Purpose**: To make it easier to create beautiful and informative statistical graphics.

## Why Use Seaborn Over Matplotlib?

1. **Simpler Syntax**:
    - Seaborn requires less code to create complex visualizations.
    - Example: A line plot with Seaborn needs fewer lines of code than with Matplotlib.

2. **Better Default Styles**:
    - Seaborn comes with attractive default styles.
    - This means you get good-looking plots without much customization.

3. **Built-in Themes**:
    - Seaborn offers themes that can change the appearance of all plots easily.
    - Example: `sns.set_style("whitegrid")` applies a grid background to all plots.

4. **Integrated with Pandas**:
    - Seaborn works well with Pandas DataFrames.
    - This makes it easy to plot data directly from DataFrames.

5. **Advanced Statistical Plots**:
    - Seaborn provides functions for complex visualizations like heatmaps, violin plots, and pair plots.
    - These are either not available or harder to create in Matplotlib.

6. **Ease of Plot Customization**:
    - Seaborn simplifies the customization of plots.
    - Example: Changing colors, adding labels, and adjusting plot elements are more straightforward.

```python
df = sns.load_dataset('tips')
```

```python
df.sample(10)
```

### a. Scatterplot -> axes level function.

```python
sns.scatterplot(data = df , x = 'total_bill' , y = 'tip')
```

```python
sns.scatterplot(data = df , x = 'total_bill' , y = 'tip' , hue = 'sex')
```

```python
sns.scatterplot(data = df , x = 'total_bill' , y = 'tip' , style = 'time')
```

```python
sns.scatterplot(data = df , x = 'total_bill' , y = 'tip' , size = 'size')
```

```python
sns.scatterplot(data = df, x = 'total_bill', y = 'tip', hue = 'sex', style = 'time', size = 'size')
```

## relplot -> figure level function
 - scatterplot

```python
sns.relplot(kind = 'scatter' , data = df, x = 'total_bill', y = 'tip', hue = 'sex', style = 'time', size = 'size')
```
