## Seaborn: Matrix Plots

### What are Matrix Plots?

- **Matrix Plots**: Visualizations that show data in a grid format.
- **Each cell in the grid** represents a data point.

### Why Use Matrix Plots?

- To easily see **patterns, relationships, and anomalies** in data.
- Useful for **comparing multiple variables**.

### What Do They Show?

- **Heatmaps**: Color-coded representation of data.
- **Cluster Maps**: Heatmaps with hierarchical clustering.

### When to Use Matrix Plots?

- When you need to **visualize relationships** between multiple variables.
- For **correlation matrices**, **feature comparisons**, and **pattern recognition**.


```python
import seaborn as sns
import plotly.express as px
```

## Type of data :
1. Numerical data
2. Categorical data.

```python
country_df = px.data.gapminder()
```

```python
country_df
```

## Formatting data.
 - Want to plot a graph of lifeExp of all the countries of continent with respect to year.

```python
temp_df = country_df[country_df['continent'] == 'Asia'].pivot(index = 'country' , columns = 'year' , values = 'lifeExp')
```

```python
temp_df
```

```python
import matplotlib.pyplot as plt
```

## Heatmap()

```python
plt.figure(figsize=(10,8))
sns.heatmap(temp_df)
```

## Parameters ->
1. annot = True

```python
plt.figure(figsize=(10,8))
sns.heatmap(temp_df , annot = True)
```

## Parameter->
2. linewidth

```python
plt.figure(figsize=(10,8))
sns.heatmap(temp_df,annot = True,linewidth = 0.5)
```

## Parameter ->
3. cmap = 'viridis' or 'summer' or 'winter' or 'jet'

```python
plt.figure(figsize=(10,8))
sns.heatmap(temp_df,annot = True,linewidth = 0.5,cmap = 'jet')
```
