# Categorical Plots

#### Figure level function: catplot()
#### Axis level functions: 
1. stripplot 
2. swarmplot
3. boxplot
4. violinplot
5. barplot
6. countplot
7. pointplot

```python
import seaborn as sns
```

```python
tips= sns.load_dataset("tips")
```

```python
tips.head()
```

## Categorical scatter plot
1. stripplot
2. swarmplot

```python
# 1. 
sns.stripplot(data=tips, x='day', y='total_bill')
```

```python
# jitter
sns.stripplot(data=tips, x='day', y='total_bill',jitter=False)
```

```python
# hue
sns.stripplot(data=tips, x='day', y='total_bill',hue='sex')
```

```python
# figure level function -> catplot
sns.catplot(data=tips, x='day', y='total_bill',kind="strip")
```

```python
sns.catplot(data=tips, x='day', y='total_bill',kind="strip",jitter=0.3)
```

```python
# hue
sns.catplot(data=tips, x='day', y='total_bill',kind="strip",hue='sex')
```

2. swarmplot -> shows the distribution of data more clear 

```python
sns.swarmplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm',hue='sex')
```

## Categorical Distribution plot
1. Boxplot
2. Violinplot


```python
sns.boxplot(data=tips,x='sex',y='total_bill')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
#hue
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
# univariate analysis using boxplot
sns.boxplot(data=tips,y='total_bill')
```

```python
# using figure level function
sns.catplot(data=tips, x='day', y='total_bill',hue='sex',kind='box')
```

2. Violinplot -> (boxplot+KDEplot)

```python
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
# hue
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
# split
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True)
```

```python
# violinplot using figure level catplot
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',split=True,kind='violin')
```

## Categorical estimate plots
1. barplot
2. pointplot
2. countplot

1. barplot()

-> The upper black lines are known as errorbar

```python
sns.barplot(data=tips,x='sex',y='total_bill')
```

```python
# hue
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker')
```

```python
# errorbar
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',errorbar=None)
```

```python
# figure level of barplot
sns.catplot(data=tips,x='sex',y='total_bill',hue='smoker',kind='bar')
```

2. pointplot()    
-> It just remove the bars from bar plot and connect the top most points of the bars.

```python
sns.pointplot(data=tips,x='sex',y='total_bill',errorbar=None)
```

```python
# figure level of pointplot
sns.catplot(data=tips,x='sex',y='total_bill',hue='smoker',errorbar=None,kind='point')
```

3. countplot()

```python
sns.countplot(data=tips,x='sex')
```

```python
# hue
sns.countplot(data=tips,x='sex',hue='day')
```

```python
# figure level of countplot()
sns.catplot(data=tips,x='sex',hue='day',kind='count')
```

## faceting using catplot()   
-> more than one graph in a single figure.

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='box',col='smoker',row='time')
```
