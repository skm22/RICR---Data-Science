```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.info()
```

```python
# Axis-level
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
plt.title("Body Mass vs. Flipper Length (by Species)")
plt.show()

# Figure-level
sns.relplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species", kind="scatter")
```

```python
# Axis-level
sns.lineplot(data=penguins, x="island", y="body_mass_g", hue="species", ci="sd")
plt.title("Trend of Body Mass by Island")
plt.show()

# Figure-level
sns.relplot(data=penguins, x="island", y="body_mass_g", hue="species", kind="line", ci="sd")
```

```python
# Axis-level
sns.barplot(data=penguins, x="species", y="body_mass_g", ci="sd")
plt.title("Mean Body Mass Across Species")
plt.show()

# Figure-level
sns.catplot(data=penguins, x="species", y="body_mass_g", kind="bar", ci="sd")
```

```python
# Axis-level
sns.boxplot(data=penguins, x="species", y="flipper_length_mm", hue="sex")
plt.title("Box Plot: Flipper Length by Species and Gender")
plt.show()

# Figure-level
sns.catplot(data=penguins, x="species", y="flipper_length_mm", hue="sex", kind="box")
```

```python
# Axis-level
sns.violinplot(data=penguins, x="species", y="flipper_length_mm", hue="sex", split=True)
plt.title("Violin Plot: Flipper Length by Species and Gender")
plt.show()

# Figure-level
sns.catplot(data=penguins, x="species", y="flipper_length_mm", hue="sex", kind="violin", split=True)
```

```python
# Axis-level
sns.countplot(data=penguins, x="species", hue="island")
plt.title("Count of Penguins by Species and Island")
plt.show()

# Figure-level
sns.catplot(data=penguins, x="species", hue="island", kind="count")
```

```python
# Compute correlation matrix
corr = penguins.corr(numeric_only=True)

# Axis-level
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# With Subplots
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", ax=ax)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(penguins, hue="species", diag_kind="kde")
plt.show()
```

```python
# Axis-level
sns.regplot(data=penguins, x="flipper_length_mm", y="body_mass_g", scatter_kws={"alpha":0.5})
plt.title("Regression Plot: Body Mass vs. Flipper Length")
plt.show()

# Figure-level
sns.lmplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species", ci=95)
```

```python
sns.residplot(data=penguins, x="bill_length_mm", y="body_mass_g")
plt.title("Residual Plot: Bill Length vs. Body Mass")
plt.show()
```

```python
# Histogram (Axis-level)
sns.histplot(data=penguins, x="body_mass_g", hue="species", multiple="stack", kde=True)
plt.title("Histogram: Body Mass Distribution Across Species")
plt.show()

# Histogram (Figure-level)
sns.displot(data=penguins, x="body_mass_g", hue="species", multiple="stack", kde=True)
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", kind="scatter")
plt.show()

# KDE Joint Plot
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", kind="kde")
plt.show()
```
