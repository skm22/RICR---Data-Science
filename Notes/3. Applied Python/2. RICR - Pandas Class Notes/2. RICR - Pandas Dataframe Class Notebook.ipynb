### 18. apply()

```python
New_data
```

```python
def percentage(row):
    cgpa = row['cgpa']
    a = cgpa * 10
    return str(a) + '%'
```

### 19. isin()

### 20. corr()
-> correlation. 

Parameters: 

-> method : 
 - pearson: standard correlation coefficient 
 - kendall: Kendall Tau correlation coefficient 
 - spearman: Spearman rank correlation        

-> numeric_only : Whether only the numeric values are to be operated upon or not. It is set to False by default.

```python
New_data.corr(numeric_only=True)
```

```python
New_data.corr(method='pearson',numeric_only=True)
```

```python
New_data.corr(method='kendall',numeric_only=True)
```

```python
New_data.corr(method='spearman',numeric_only=True)
```
