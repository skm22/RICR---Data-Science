# Histogram in Matplotlib


---



---


## What is it?
A histogram is a type of bar graph that represents the distribution of a dataset by grouping data into bins and plotting the frequency of data points in each bin.


---


## Why do we use it?
 - To visualize the distribution of a continuous dataset.
 - To understand the underlying frequency distribution (e.g., normal distribution, skewed distribution).


---


## Purpose :
 - To provide a visual representation of the distribution of data points in a dataset.
 - To identify patterns, central tendencies, and variability in the data.


---


## What it tells us
 - The frequency of data points within specified ranges (bins).
 - The shape of the data distribution (e.g., symmetric, skewed).
 - The spread and central tendency of the data.


---


## When to Use
1. **Understanding Distribution**: When you need to understand the distribution of a single variable.
- Scenario: Visualizing the distribution of exam scores to see how students performed.
2. **Identifying Patterns**: When you want to identify patterns such as skewness, modality, or outliers.
- Scenario: Observing the distribution of ages in a population to identify age groups with higher frequencies.
3. **Comparing Distributions**: When comparing the distributions of the same variable across different groups.
- Scenario: Comparing the distribution of test scores between two different classes.

```python
import matplotlib.pyplot as plt
```

## Type of Data
 - Numerical Data: Continuous data that can be grouped into ranges.
### Examples:
1. Exam scores of students.
2. Heights of individuals.
3. Daily temperatures.

```python
numbers = [16, 32, 17, 25, 22, 19, 33, 12, 22, 10, 32, 14, 16, 40, 40, 39, 13, 21, 11, 25, 23, 36, 12, 22, 17, 21,
 29, 25, 34, 32, 20, 22, 22, 32, 10, 37, 20, 27, 16, 37, 35, 13, 40 , 16, 25, 17, 29, 17, 14, 30, 25,
 11, 34, 11, 16, 23, 13, 24, 17, 13, 25, 17, 19]
```

## hist()

```python
plt.hist(numbers)
plt.show()
```

## Title , xlabel , ylabel

```python
plt.hist(numbers)
plt.title("Frequenct count")
plt.ylabel("Count")
plt.xlabel("Numbers")
plt.show()
```

## Color , Edgecolor

```python
plt.hist(numbers,color='m',edgecolor='r')
plt.title("Frequenct count")
plt.ylabel("Count")
plt.xlabel("Numbers")
plt.show()
```

## Count of bins

```python
plt.hist(numbers,color='m',edgecolor='r',bins=15)
plt.title("Frequenct count")
plt.ylabel("Count")
plt.xlabel("Numbers")
plt.show()
```

## Interval of bins

```python
L = [10,14,18,22,26,30,34,38,42]
plt.hist(numbers,color='m',edgecolor='r',bins = L)
plt.title("Frequenct count")
plt.ylabel("Count")
plt.xlabel("Numbers")
plt.show()
```

## range of x-axis.

```python
plt.hist(numbers,bins="auto",range=(0,100),color='m',edgecolor='r')
plt.title("Frequenct count")
plt.ylabel("Count")
plt.xlabel("Numbers")
plt.show()
```

## Cumulative = -1 : This sorts the bar in the decresing order.

```python
plt.hist(numbers,color='m',edgecolor='r',cumulative=-1)
plt.title("Frequenct count")
plt.ylabel("Count")
plt.xlabel("Numbers")
plt.show()
```

## Bottom

```python
plt.hist(numbers,color='m',edgecolor='r',bottom = 10)
plt.title("Frequenct count")
plt.ylabel("Count")
plt.xlabel("Numbers")
plt.show()
```

## Align = 'left' or 'rigth' or 'mid'

```python
plt.hist(numbers,color='m',edgecolor='r',align='left')
plt.title("Frequenct count")
plt.ylabel("Count")
plt.xlabel("Numbers")
plt.show()
```

## histtype = 'step' or 'stepfilled'

```python
plt.hist(numbers,histtype='step')
plt.title("Frequenct count")
plt.ylabel("Count")
plt.xlabel("Numbers")
plt.show()
```

## orientation = 'horizontal' or 'vertical'

```python
plt.hist(numbers,color='m',edgecolor='r',orientation='horizontal')
plt.title("Frequenct count")
plt.xlabel("Count")
plt.ylabel("Numbers")
plt.show()
```

## width of bars

```python
plt.hist(numbers,color='m',edgecolor='r',width=2)
plt.title("Frequenct count")
plt.xlabel("Count")
plt.ylabel("Numbers")
plt.show()
```

## rwidth : related width of bars.

```python
plt.hist(numbers,color='m',edgecolor='r',rwidth=0.8)
plt.title("Frequenct count")
plt.xlabel("Count")
plt.ylabel("Numbers")
plt.show()
```

## log = True

```python
plt.hist(numbers,color='m',edgecolor='r',log=True)
plt.title("Frequenct count")
plt.xlabel("Count")
plt.ylabel("Numbers")
plt.show()
```

## label

```python
plt.hist(numbers,color='m',edgecolor='r',label='Numbers')
plt.title("Frequenct count")
plt.xlabel("Count")
plt.ylabel("Numbers")
plt.legend()
plt.show()
```

## axvline()

```python
import numpy as np
```

```python
plt.hist(numbers,color='m',edgecolor='r',label='numbers')
plt.title("Frequenct count")
plt.ylabel("Count")
plt.xlabel("Numbers")
plt.axvline(np.mean(numbers),color='g',label='avg-line')
plt.legend()
plt.show()
```

## All the customizations in one graph.

```python
L = [10,14,18,22,26,30,34,38,42]
plt.hist(numbers,bins=L,color='orange',edgecolor='r',align='left',histtype='stepfilled',rwidth=0.8,orientation='vertical',label = 'Numbers')
plt.title("Frequenct count")
plt.ylabel("Count")
plt.xlabel("Numbers")
plt.axvline(np.mean(numbers),color='g',label='avg-line')
plt.legend()
plt.show()
```
