# Box Plot in Matplotlib

Box-Plot-and-Whisker-Plot-1.png



**What is it?**
- A box plot, also known as a whisker plot, is a chart that shows the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.

**Why do we use it?**
- To understand the spread and skewness of data.
- To easily identify outliers.

**Purpose**
- To provide a clear summary of the distribution and variability of the data.
- To compare distributions between multiple datasets.

**What it tells us**
- **Minimum**: The smallest data point, excluding outliers.
- **First Quartile (Q1)**: The median of the lower half of the dataset (25th percentile).
- **Median (Q2)**: The middle value of the dataset (50th percentile).
- **Third Quartile (Q3)**: The median of the upper half of the dataset (75th percentile).
- **Maximum**: The largest data point, excluding outliers.
- **Outliers**: Data points that are significantly higher or lower than the rest of the data.


**When to Use**
- **Comparing Distributions**: When you want to compare the distribution of data across different groups.
  - **Scenario**: Comparing test scores of students from different classes.
- **Identifying Outliers**: When you need to find out if there are any extreme values in your data.
  - **Scenario**: Checking for unusual sales figures in a monthly sales dataset.
- **Understanding Spread and Skewness**: When you want to see the spread and skewness of your data.
  - **Scenario**: Analyzing the distribution of heights in a population to see if it is skewed.


```python
import matplotlib.pyplot as plt
```

## Type of Data
- **Numerical Data**: Data that is continuous and can be ordered.
- **Examples**:
  - Test scores of students.
  - Heights of individuals.
  - Monthly sales figures.

```python
x = [10,20,30,40,50,60,70]
```

## plt.boxplot()

```python
plt.boxplot(x)
plt.show()
```

## **notch**: Adds a small notch to the median line of the boxplot, giving an indication of the confidence interval around the median.

```python
plt.boxplot(x,notch=True)
plt.show()
```

## **vert**: Specifies whether the boxplot is oriented vertically (True) or horizontally (False).

```python
plt.boxplot(x,vert=False)
plt.show()
```

## **widths**: Sets the width of the boxes in the boxplot.

```python
plt.boxplot(x,widths = 0.5)
plt.show()
```

## **labels**: Adds labels to the boxplot, typically used to label different categories or groups.

```python
plt.boxplot(x,labels = ['python'])
plt.show()
```

## **patch_artist**: Allows filling the box of the boxplot with a color.

```python
plt.boxplot(x,patch_artist=True)
plt.show()
```

## **showmeans**: Displays a marker for the mean value inside the box.

```python
plt.boxplot(x,showmeans = True)
plt.show()
```

## Outlier

```python
y = [10,20,30,40,50,60,110]
plt.boxplot(y)
plt.show()
```

## **whis** : connect the outlier to the whiskers.

```python
plt.boxplot(y,whis=2)
plt.show()
```

## sym: Specifies the appearance of outlier points in the boxplot.
 - By default, outliers are represented as points beyond the whiskers.
 - You can customize this parameter to change how outliers are displayed, for example, by setting sym='o' to show outliers as circles, or sym='+' to show them as plus signs, sym = 'g' chnages the color to the outliers etc.

```python
plt.boxplot(y,sym='g+')
plt.show()
```

## Customization Properties
- **whiskerprops**: Customizes the appearance of the whiskers (lines extending from the box).
- **capprops**: Customizes the appearance of the caps (the ends of the whiskers).
- **boxprops**: Customizes the appearance of the boxes in the boxplot.
- **medianprops**: Customizes the appearance of the median line inside the box.

```python
plt.boxplot(x,boxprops = {'color': 'r'},capprops = {'color':'g'},whiskerprops = {'color':'b'},medianprops={'color':'black'})
plt.show()
```

## Mutiple Box plot.

```python
x1 = [10,20,30,40,50,60,70,120]
x2 = [40,50,50,60,40,70,10,1,90]
y = [x1,x2]
plt.boxplot(y,labels = ['pyhon','C++'])
plt.show()
```

## Combined all the customizations for the box plot in one chart.

```python
import matplotlib.pyplot as plt

# Data to plot
x = [10, 20, 30, 40, 15, 60, 70, 80, 80, 90, 110]
labels = ["Python"]

boxprops = dict(color="blue", linewidth=2)  # Customize the box properties
flierprops = dict(marker='o', color='red', markersize=8)  # Customize the outliers
medianprops = dict(color='green', linewidth=2)  # Customize the median line
meanprops = dict(marker='D', markerfacecolor='orange', markersize=10)  # Customize the mean marker

# Create a boxplot
plt.boxplot(x,
            notch=False,  # Add a notch to the box
            vert=True,  # Vertical box alignment
            patch_artist=True,  # Fill the box with color
            widths=0.4,  # Width of the box
            labels=labels,  # Labels for the box
            showmeans=True,  # Show the mean marker
            showcaps=True,  # Show the caps at the end of the whiskers
            showbox=True,  # Show the box
            whiskerprops=dict(color="purple", linewidth=2),  # Customize whiskers
            capprops=dict(color="black", linewidth=2),  # Customize caps
            boxprops=boxprops,  # Apply custom box properties
            flierprops=flierprops,  # Apply custom outlier properties
            medianprops=medianprops,  # Apply custom median properties
            meanprops=meanprops  # Apply custom mean properties
           )

# Add a title and labels
plt.title('Boxplot of Python Data')
plt.xlabel('Programming Language')
plt.ylabel('Values')

# Display the chart
plt.show()
```
