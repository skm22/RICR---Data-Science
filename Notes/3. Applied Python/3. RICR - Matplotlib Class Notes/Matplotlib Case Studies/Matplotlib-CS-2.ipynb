```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic Dataset.csv")
```

```python
df.head(3)
```

```python
df.info()
```

### What is the age distribution of passengers on the Titanic?

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=30, color='purple', edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

### How does the survival rate vary across different passenger classes? 

```python
survival_rate = df.groupby('Pclass')['Survived'].mean()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(survival_rate.index, survival_rate, color='lightpink', edgecolor='black')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

### What is the distribution of survivors and non-survivors?

```python
survived_counts = df['Survived'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
plt.bar([0, 1], survived_counts, color=['skyblue', 'salmon'], edgecolor = 'black')
plt.title('Number of Survivors vs Non-Survivors')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0, 1], ['Non-Survived', 'Survived'], rotation=0)
plt.show()
```

### What is the proportion of passengers in each class?

```python
pclass_counts = df['Pclass'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(pclass_counts, labels=pclass_counts.index, autopct='%1.1f%%',explode=[0.0,0.0,0.2])
plt.title('Proportion of Passenger Classes')
plt.show()
```

### Is there a relationship between a passenger's age and the fare they paid?

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Fare'], color='red', alpha=0.6)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

### How does the age distribution differ between survivors and non-survivors?

```python
plt.figure(figsize=(8, 6))
df.boxplot(column='Age', by='Survived', grid=False, patch_artist=True,
           boxprops=dict(facecolor='lightblue', color='blue'),
           medianprops=dict(color='red'))
plt.title('Age Distribution by Survival Status')
plt.suptitle('')
plt.xlabel('Survival')
plt.ylabel('Age')
plt.xticks([1, 2], ['Non-Survived', 'Survived'])
plt.show()
```
