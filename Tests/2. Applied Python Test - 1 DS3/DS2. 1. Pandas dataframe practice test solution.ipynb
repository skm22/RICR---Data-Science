## Solution file of pandas practice test

```python
import seaborn as sns
penguins = sns.load_dataset('penguins')
```

```python
penguins.sample(4)
```

```python
penguins.info()
```

### 1. Replace all missing values in the 'bill_depth_mm' column with the value that occurs most frequently in it.

```python
most_frequent_bill_depth = penguins['bill_depth_mm'].mode()[0]
penguins['bill_depth_mm'].fillna(most_frequent_bill_depth)
penguins['bill_depth_mm'].isnull().sum()    
```

### 2. Retrieve rows where penguins have an 'island' value of 'Torgersen' and only display their 'flipper_length_mm' and 'body_mass_g'.

```python
torgersen_penguins = penguins[penguins['island'] == 'Torgersen'][['flipper_length_mm', 'body_mass_g']]
torgersen_penguins
```

### 3. Filter penguins from the 'Dream' island that are male. Sort the result by 'body_mass_g' in descending order.

```python
dream_males = penguins[(penguins['island'] == 'Dream') & (penguins['sex'] == 'Male')].sort_values('body_mass_g', ascending=False)
dream_males
```

### 4. Select the first 10 rows of penguins where 'body_mass_g' is greater than the median body mass of all penguins.

```python
median_body_mass = penguins['body_mass_g'].median()
penguins[penguins['body_mass_g'] > median_body_mass].head(10)
```

### 5. Find rows where penguins’ 'bill_length_mm' is either below 35 or above 50, and display their 'species' and 'sex'.

```python
penguins[(penguins['bill_length_mm'] < 35) | (penguins['bill_length_mm'] > 50)][['species', 'sex']]
```

### 6. Identify duplicate entries in the dataset based on the 'species' and 'island' columns, and remove them permanently.

```python
penguins.drop_duplicates(subset=['species', 'island'])    
```

### 7. Get rows for the first 5 penguins with a 'body_mass_g' greater than 5000 and sort the result by 'bill_length_mm'.

```python
penguins[penguins['body_mass_g'] > 5000].sort_values('bill_length_mm').head(5) 
```

### 9. Check if any column in the dataset contains missing values. If yes, display the column names.

```python
penguins.columns[penguins.isnull().any()]
```

### 10. Filter penguins whose 'flipper_length_mm' is greater than 200 and 'bill_depth_mm' is less than 18. Show only their 'species', 'bill_length_mm', and 'flipper_length_mm'.

```python
penguins[(penguins['flipper_length_mm'] > 200) & (penguins['bill_depth_mm'] < 18)][['species', 'bill_length_mm', 'flipper_length_mm']]
```

### 11. Assign ranks to penguins based on their 'body_mass_g', with higher body masses getting higher ranks. Allow ties in the ranking.

```python
penguins['body_mass_rank'] = penguins['body_mass_g'].rank(ascending=False)
penguins[['body_mass_g', 'body_mass_rank']].head(10)  
```

### 12. Retrieve penguins whose 'flipper_length_mm' is greater than 200 and 'sex' is not null.

```python
penguins[(penguins['flipper_length_mm'] > 200) & (penguins['sex'].notnull())] 
```

### 13. Find the rows where the 'sex' column is null, and extract the 'species' and 'island' columns for those penguins.

```python
penguins[penguins['sex'].isnull()][['species', 'island']]    
```

### 14. Create a new column named 'Above_Avg_Mass' that stores True if a penguin's 'body_mass_g' is greater than the dataset's average body mass, otherwise False.

```python
average_body_mass = penguins['body_mass_g'].mean()
penguins['Above_Avg_Mass'] = penguins['body_mass_g'] > average_body_mass
penguins[['body_mass_g', 'Above_Avg_Mass']].head()
```

### 15. Extract penguins whose 'island' is either 'Biscoe' or 'Dream' and store their information in a new DataFrame.

```python
penguins[penguins['island'].isin(['Biscoe', 'Dream'])]
```
